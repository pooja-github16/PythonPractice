```python
#Temprature conversion  convert celsius into fahrenheit and vice-versa

def convert_temp(temp,unit):
  if unit=='c':
    return (temp *9/5 + 32) #celsius to fahrenheit
  elif unit== 'f':
    return (temp-32)*5/9
  else:
    return None

print(convert_temp(20,'c'))
print(convert_temp(68,'f'))
```

```python
#check password is strong or not
def is_strong_pass(password):
  if len(password)<8:
    return False
  if not any (char.isdigit() for char in password):
    return False
  if not any(char.islower() for char in password) and any(char.isupper() for char in password):
    return False
  if not any(char in '!@#$%^&*' for char in password):
    return False
  return True

print(is_strong_pass('828282'))

```

```python
#checking string is palindrome

def palindrome(str):
  str=str.lower().replace(" ","")
  return str==str[::-1]


print(palindrome('abcdf'))
print(palindrome('noon civic noon'))
```

```python
#code factorial of number

def factorial(n):
  if n>0:
    fact=1
    for i in range(1,n+1):
      fact*=i
    return fact
  else:
    return 'Number must be greater than 0'


#calling
print(factorial(5))
```

```python
#Lambda function syntax: lambda arguments: expression

addition= lambda a,b: a+b

#calling function
print(addition(10,20))

#find even/odd no

is_even=lambda n:n%2==0
print(is_even(7))
```

```python
#Map function :The map() function in Python is a built-in function used to apply a specified function to each item
#in an iterable (or multiple iterables) and return an iterator that yields the results.


def sequre(n):
  return n*n
number=[1,2,3,4,5,6,7,8]

#calling function in map
print(list (map(sequre,number)))

#map with lambda function return cube
print(list(map(lambda x:x*x*x, number)))

#add two list using map and lambda

list1=[1,2,3,4,5]
list2=[6,7,8,9,1]

added_list=list(map(lambda x,y: x+y, list1,list2))
print(f" addition of two list : {added_list}")

#return all the list in uper case
fruits=['apple','orange','grape','pineapple']

print(list(map(str.upper,fruits)))

#Using lambda
upper_case=(list(map(lambda x:x.upper(), fruits)))
print(upper_case)
```

```python
#Filter function
list1=[2,3,4,5,7,8,7]

result= list(filter(lambda x:x>5, list1))
print(result)

#filter records

person=[
    {'name': 'Asha', 'age': 35},
    {'name': 'Neha', 'age': 30}
]

def age_filter(person):
  return person['age']>20

newlist=list(filter(age_filter, person))
print(newlist)
```
