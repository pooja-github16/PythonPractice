```python
'''#Question 2

You are given a list of integers representing the IDs of students who have signed up for an event, but some students have signed up multiple times by mistake. Your task is to:

Remove duplicates to get the unique IDs of students who signed up. Identify students who have both signed up and attended the event (IDs are given in a separate list of attendees). Classify each student as: "Attended" if they signed up and attended. "No-show" if they signed up but did not attend. Finally, print a summary showing the count of "Attended" and "No-show" students. Input:

A list of integers signups, where each element represents a student ID who signed up for the event. (It may contain duplicates.) A list of integers attendees, where each element represents a student ID who attended the event. Output: Print each unique student ID along with its status ("Attended" or "No-show") and a summary like the example below.

# Example Input:

signups = [101, 102, 103, 104, 101, 105, 106, 102)

attendees [102, 104, 106,107]'''

signups = [101, 102, 103, 104, 101, 105, 106, 102]

attendees=[102,105,107]

#Remove duplicate and find unique id: Set for unique value
unique_signups=set(signups)
unique_attendess= set(attendees)

#create counter for attended and no-show student
attended_count=0
no_show_count=0

#classify them
for i in unique_signups:
  if i in unique_attendess:
    print(i,": Attended")
    attended_count+=1
  else:
    print(i,': No show.')
    no_show_count+=1

#Print summery
print()
print("Summery")
print("Number of Attenened student ",attended_count)
print("Number of No-show studenr ",no_show_count)





```

```python
#You are given two lists of integer : find  the unique common elent if not print msg "No common elements"

list1=[1,2,3,4,5,6,7]
list2=[5,7,8,9,10]

#converting list to set to get unique values
set1=set(list1)
set2=set(list2)

#find common elements
common=set1 & set2

#check and print the result
if common:
  print("sorted common list:",sorted(common))
else:
  print("No common element.")
```

```python
'''You are given two list
cart A: list of items purchesed by user(May contain dublicate)
Inventory :A list of items available in stocks
Program to Identify and print item purchesed by user that are not in stocks, count the nuber of  unique item purchesed by user.

Input:
cart = ['apple','orange','banana','grape','banana']
inventory=['apple','banana','kiwi','grape']
output:
out of stock item = 'orange'
unique item purchesed = 4 '''

cart = ['apple','orange','banana','grape','banana']
inventory=['apple','banana','kiwi','grape']

#convert to set to get unique value
unique_cart=set(cart)
unique_inventory= set(inventory)

#identify out of stock elements
out_of_stock = unique_cart - unique_inventory
print(out_of_stock)

#count unique item purchesed
print("Number of Unique items purchesed ",len(unique_cart)-len(out_of_stock))

```

```python
''' you are given a dictionary where the keys are the student names and values of the list of there score
in three subjects. calculate and print the average score of each student.'''

gread = {
    'Alice': {90,95,98},
    'Bob': {70,80,75},
    'John': {75,70,80}
}

#calculate average score
for i,j in gread.items():
  avg = sum(j)/len(j)
  print(i, ':', round(avg,2))
  print(i, ':', f"{avg:.2f}") #fatmat decimal in 2 place both way can be use
```
